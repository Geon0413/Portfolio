```python
import pandas as pd
```

```python
sub=pd.read_csv('preproc_data\preprocessed_subscription_log.csv')
```

```python
sub.head()
```

```python
del sub['Unnamed: 0']
```

```python
sub.head()
```

```python
sub['end_date']=pd.to_datetime(sub['end_date'])
```

```python
sub['created_at']=pd.to_datetime(sub['created_at'])
```

```python
sub['month']=sub['end_date']-sub['created_at']
```

```python
sub.head()
```

```python
test= sub.sort_values(['user_id','created_at'])
```

```python
test.head(20)
```

```python
test.duplicated(subset='user_id',keep=False)
```

```python
df1=test.groupby(['user_id'],as_index=False)[['month']].sum()
```

```python
df2=test.groupby(['user_id'],as_index=False)[['price']].sum()
```

```python
df3=test.groupby(['user_id'],as_index=False)[['refund']].sum()
```

```python
df_mer=pd.merge(df1,df2)
```

```python
df_mer=pd.merge(df_mer,df3)
```

```python
df_mer.rename(columns={'month':'period_use'},inplace=True)
```

```python
df_mer.rename(columns={'price':'total_fee'},inplace=True)
```

```python
df_mer.head()
```

```python
tt2=pd.get_dummies(test)
```

```python
tt2.head()
```

```python
tt2[['user_id','target_customer_personal','target_customer_smallBiz','target_customer_middleBiz','target_customer_bigBiz']]
```

```python
df11=tt2.groupby('user_id',as_index=False)[['target_customer_personal']].sum()
```

```python
df12=tt2.groupby('user_id',as_index=False)[['target_customer_smallBiz']].sum()
```

```python
df13=tt2.groupby('user_id',as_index=False)[['target_customer_middleBiz']].sum()
```

```python
df14=tt2.groupby('user_id',as_index=False)[['target_customer_bigBiz']].sum()
```

```python
df_mer1= pd.merge(df11,df12)
```

```python
df_mer1= pd.merge(df_mer1,df13)
```

```python
df_mer1= pd.merge(df_mer1,df14)
```

```python
df_mer1.rename(columns={'target_customer_personal':'personal','target_customer_smallBiz':'smallBiz','target_customer_middleBiz':'middleBiz','target_customer_bigBiz':'bigBiz'},inplace=True)
```

```python
sub_mer=pd.merge(df_mer,df_mer1)
```

```python
sub_mer=sub_mer.rename(columns={'period_use':'period_sub'})
```

```python
sub_mer.to_csv('sub_total.csv')
```

```python
df_user= pd.read_csv('user_total.csv')
```

```python
sub_user=pd.merge(sub_mer, df_user, how='left', left_on='user_id', right_on='user_id')
```
