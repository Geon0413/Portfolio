```python
import pandas as pd
```

```python
sfx =pd.read_csv('sfx_down_total.csv')
sub_total= pd.read_csv('sub_user_total.csv')
song= pd.read_csv('song_down_total.csv')
etc=pd.read_csv('etc_down_total.csv')
userid= pd.read_csv('test_userid_type.csv')
genre= pd.read_csv('favorite_genre.csv')
```

```python
del sub_total['Unnamed: 0']
del sfx['Unnamed: 0']
del song['Unnamed: 0']
del etc['Unnamed: 0']
```

```python
df1=pd.merge(sub_total, sfx, how='left', left_on='user_id', right_on='user_id')
```

```python
df2=pd.merge(df1, song, how='left', left_on='user_id', right_on='user_id')
```

```python
df3=pd.merge(df2, etc, how='left', left_on='user_id', right_on='user_id')
```
